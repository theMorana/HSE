```python
# 2.1 Создание самого простого графика
# Основная структура: plt.figure() → plt.plot() → plt.show()

# Генерация простых данных
x = np.linspace(0, 10, 100)  # 100 точек от 0 до 10
y = np.sin(x)  # Синусоида

# Создание фигуры с указанием размера
plt.figure(figsize=(8, 4))

# Построение линии графика с параметрами:
# x, y - данные
# 'b-' - стиль линии: b=blue, -=сплошная линия
# linewidth - толщина линии
# label - метка для легенды
plt.plot(x, y, 'b-', linewidth=2, label='sin(x)')

# Добавление заголовка
plt.title('Простой линейный график', fontsize=14)

# Подписи осей (обязательны для читаемости)
plt.xlabel('X - значение')
plt.ylabel('Y - значение')

# Включение сетки
plt.grid(True, alpha=0.3)  # alpha - прозрачность

# Добавление легенды
plt.legend()

# Отображение графика
plt.show()
```

# 3. Типы графиков и их применение

```python
# 3.1 Создаем фигуру с разными типами графиков
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
fig.suptitle('Различные типы графиков и их применение', fontsize=16)

# Данные для примеров
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D']
values1 = [10, 25, 40, 15]
values2 = [15, 20, 35, 20]

# 1. Линейный график (тренды, временные ряды)
ax = axes[0, 0]
ax.plot(x, y1, 'b-', linewidth=2)
ax.set_xlabel('Время / Последовательность')
ax.set_ylabel('Значение')
ax.set_title('Линейный график\n(тренды, временные ряды)')
ax.grid(True, alpha=0.3)

# 2. Scatter plot (корреляции, кластеры)
ax = axes[0, 1]
np.random.seed(42)
x_scatter = np.random.randn(100)
y_scatter = x_scatter * 2 + np.random.randn(100) * 0.5
ax.scatter(x_scatter, y_scatter, alpha=0.6, c=np.abs(x_scatter),
           cmap='viridis', edgecolors='black', linewidth=0.5)
# c - цветовая кодировка по значению
# cmap - цветовая карта
ax.set_xlabel('Признак X')
ax.set_ylabel('Признак Y')
ax.set_title('Scatter plot\n(корреляции, кластеры)')

# 3. Гистограмма (распределения)
ax = axes[0, 2]
data_hist = np.concatenate([
    np.random.normal(0, 1, 500),
    np.random.normal(5, 1.5, 300)
])
ax.hist(data_hist, bins=40, alpha=0.7, color='lightcoral', edgecolor='black')
ax.set_xlabel('Значение')
ax.set_ylabel('Частота')
ax.set_title('Гистограмма\n(распределение данных)')

# 4. Box plot (статистика, выбросы)
ax = axes[1, 0]
data_box = [np.random.normal(0, 1, 100),
            np.random.normal(2, 1.5, 100),
            np.random.normal(5, 1, 100)]
ax.boxplot(data_box, labels=['Группа A', 'Группа B', 'Группа C'])
# Показывает: медиану, квартили, выбросы
ax.set_ylabel('Значение')
ax.set_title('Box plot\n(статистика, сравнение групп)')
ax.grid(True, alpha=0.3)

# 5. Bar chart (сравнение категорий)
ax = axes[1, 1]
x_index = np.arange(len(categories))
width = 0.35
bars1 = ax.bar(x_index - width/2, values1, width, label='Набор 1', alpha=0.8)
bars2 = ax.bar(x_index + width/2, values2, width, label='Набор 2', alpha=0.8)
ax.set_xlabel('Категории')
ax.set_ylabel('Значения')
ax.set_title('Grouped Bar chart\n(сравнение категорий)')
ax.set_xticks(x_index)
ax.set_xticklabels(categories)
ax.legend()

# 6. Pie chart (доли, проценты)
ax = axes[1, 2]
ax.pie(values1, labels=categories, autopct='%1.1f%%', startangle=90)
# autopct - формат отображения процентов
# startangle - угол начала
ax.set_title('Круговая диаграмма\n(доли, проценты)')

plt.tight_layout()
plt.show()
```

# 4. Seaborn: статистическая визуализация

```python
# 4.1 Основные отличия seaborn от matplotlib
# - Более красивые стили по умолчанию
# - Упрощенный синтаксис для сложных графиков
# - Интеграция с pandas DataFrames
# - Встроенные статистические агрегации

# Создаем тестовый DataFrame
np.random.seed(42)
n_samples = 200
df = pd.DataFrame({
    'feature_1': np.random.normal(0, 1, n_samples),
    'feature_2': np.random.exponential(1, n_samples),
    'feature_3': np.random.uniform(-2, 2, n_samples),
    'category': np.random.choice(['A', 'B', 'C'], n_samples),
    'value': np.random.randn(n_samples)
})

# 4.2 Histogram with KDE (Kernel Density Estimation)
plt.figure(figsize=(12, 10))

# Subplot 1: Гистограмма с KDE
plt.subplot(2, 2, 1)
# kde=True - добавляет оценку плотности
# bins='auto' - автоматический подбор количества интервалов
sns.histplot(data=df, x='feature_1', kde=True, bins='auto', color='skyblue')
plt.title('Гистограмма с KDE\n(распределение + плотность)')
plt.xlabel('Feature 1')
plt.ylabel('Частота')

# Subplot 2: Box plot по категориям
plt.subplot(2, 2, 2)
# hue - разделение по категориальной переменной
sns.boxplot(data=df, x='category', y='value', palette='Set2')
plt.title('Box plot по категориям\n(сравнение распределений)')
plt.xlabel('Категория')
plt.ylabel('Значение')

# Subplot 3: Violin plot (комбинация box plot и KDE)
plt.subplot(2, 2, 3)
sns.violinplot(data=df, x='category', y='value', palette='muted', inner='quartile')
# inner='quartile' - показывает квартили внутри
plt.title('Violin plot\n(распределение + плотность)')
plt.xlabel('Категория')
plt.ylabel('Значение')

# Subplot 4: Scatter plot с регрессией
plt.subplot(2, 2, 4)
# kind='reg' - добавляет линию регрессии и доверительный интервал
sns.regplot(data=df, x='feature_1', y='feature_3', scatter_kws={'alpha':0.6})
plt.title('Scatter plot с регрессией\n(зависимость + тренд)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 3')

plt.tight_layout()
plt.show()

# 4.3 Heatmap корреляционной матрицы
plt.figure(figsize=(10, 8))
# Вычисляем корреляции между признаками
corr_matrix = df[['feature_1', 'feature_2', 'feature_3', 'value']].corr()

# annot=True - отображает значения в ячейках
# fmt='.2f' - формат чисел (2 знака после запятой)
# cmap='coolwarm' - цветовая схема
# center=0 - центр цветовой шкалы на 0
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, square=True, linewidths=1, cbar_kws={'label': 'Корреляция'})
plt.title('Тепловая карта корреляций\n(матрица зависимостей)')
plt.show()

# 4.4 Pair plot (матрица scatter plot)
plt.figure(figsize=(12, 10))
# hue - раскрашиваем по категориям
# diag_kind='kde' - на диагонали KDE вместо гистограммы
sns.pairplot(df[['feature_1', 'feature_2', 'feature_3', 'category']],
             hue='category', palette='viridis', plot_kws={'alpha': 0.6},
             diag_kind='kde')
plt.suptitle('Pair plot - матрица взаимосвязей', y=1.02)
plt.show()
```

# 5. Визуализация ML-метрик

```python
# Генерация синтетических данных обучения модели
np.random.seed(42)
epochs = np.arange(1, 51)

# Симулируем процесс обучения
train_loss = np.exp(-epochs/15) + np.random.normal(0, 0.01, len(epochs))
val_loss = np.exp(-epochs/20) + np.random.normal(0, 0.015, len(epochs))
train_acc = 1 - 0.6*np.exp(-epochs/10) + np.random.normal(0, 0.005, len(epochs))
val_acc = 1 - 0.7*np.exp(-epochs/12) + np.random.normal(0, 0.008, len(epochs))

# 5.1 Графики обучения модели
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# График потерь (loss)
ax = axes[0]
ax.plot(epochs, train_loss, 'b-', linewidth=2.5, label='Обучающая выборка')
ax.plot(epochs, val_loss, 'r--', linewidth=2.5, label='Валидационная выборка')
ax.set_xlabel('Эпоха', fontsize=12)
ax.set_ylabel('Loss', fontsize=12)
ax.set_title('График потерь (Loss) в процессе обучения', fontsize=14)
ax.legend(loc='upper right')
ax.grid(True, alpha=0.3)
# Добавляем аннотацию минимального значения
min_val_idx = np.argmin(val_loss)
ax.annotate(f'min: {val_loss[min_val_idx]:.3f}',
            xy=(epochs[min_val_idx], val_loss[min_val_idx]),
            xytext=(epochs[min_val_idx]+5, val_loss[min_val_idx]+0.05),
            arrowprops=dict(arrowstyle='->', color='red', alpha=0.7))

# График точности (accuracy)
ax = axes[1]
ax.plot(epochs, train_acc, 'g-', linewidth=2.5, label='Обучающая выборка')
ax.plot(epochs, val_acc, 'm--', linewidth=2.5, label='Валидационная выборка')
ax.set_xlabel('Эпоха', fontsize=12)
ax.set_ylabel('Accuracy', fontsize=12)
ax.set_title('График точности (Accuracy) в процессе обучения', fontsize=14)
ax.legend(loc='lower right')
ax.grid(True, alpha=0.3)
# Добавляем аннотацию максимального значения
max_val_idx = np.argmax(val_acc)
ax.annotate(f'max: {val_acc[max_val_idx]:.3f}',
            xy=(epochs[max_val_idx], val_acc[max_val_idx]),
            xytext=(epochs[max_val_idx]-15, val_acc[max_val_idx]-0.1),
            arrowprops=dict(arrowstyle='->', color='purple', alpha=0.7))

plt.tight_layout()
plt.show()

# 5.2 Confusion matrix (матрица ошибок)
plt.figure(figsize=(8, 6))

# Генерируем синтетическую confusion matrix
conf_matrix = np.array([[45, 5, 2],
                        [3, 48, 4],
                        [1, 3, 49]])

classes = ['Класс 1', 'Класс 2', 'Класс 3']

sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=classes, yticklabels=classes,
            cbar_kws={'label': 'Количество образцов'})

plt.title('Confusion Matrix\n(матрица ошибок классификации)', fontsize=14)
plt.ylabel('Истинный класс', fontsize=12)
plt.xlabel('Предсказанный класс', fontsize=12)
plt.show()

# 5.3 ROC-кривая
plt.figure(figsize=(8, 6))

# Симулируем данные для ROC-кривой
np.random.seed(42)
n_samples = 1000
y_true = np.random.randint(0, 2, n_samples)
y_scores = np.random.rand(n_samples)

from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, color='darkorange', lw=3,
         label=f'ROC кривая (AUC = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', alpha=0.5)
plt.xlim([-0.01, 1.01])
plt.ylim([-0.01, 1.01])
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC-кривая (оценка качества классификатора)', fontsize=14)
plt.legend(loc="lower right")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Создайте фигуру с 4 subplots (2x2):
# 1. Линейный график функции y = x^2 * sin(x) на интервале [0, 10]
#    Добавьте заголовок, подписи осей, сетку, легенду
# 2. Гистограмма 1000 случайных точек из нормального распределения
#    Настройте количество bins=30, добавьте заголовок
# 3. Scatter plot: 50 точек (x - равномерное распределение, y = 2x + шум)
#    Настройте размер точек, прозрачность, цвет
# 4. Bar chart: сравнение 5 разных моделей по точности
#    accuracy = [0.85, 0.92, 0.88, 0.95, 0.90]
#    Добавьте значения над столбцами

import numpy as np
import matplotlib.pyplot as plt

# Создаем фигуру с 4 subplots (2x2)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # 2 строки, 2 столбца



# 1. Линейный график y = x^2 * sin(x)
ax = axes[0, 0]
x = np.linspace(0, 10, 100)
y = x**2 * np.sin(x)

ax.plot(x, y, 'b-', linewidth=2, label='y = x^2 * sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('y = x^2 * sin(x)', fontsize=12)
ax.legend(loc='upper left')
ax.grid(True, alpha=0.3)

# 2. Гистограмма нормального распределения
ax = axes[0, 1]
data = np.random.normal(0, 1, 1000)
ax.hist(data, bins=30, alpha=0.6, color='steelblue', edgecolor='black')
ax.set_xlabel('Значение')
ax.set_ylabel('Частота')
ax.set_title('Гистограмма (1000 точек)', fontsize=12)
ax.grid(True, alpha=0.3)

# 3. Scatter plot: y = 2x + шум
ax = axes[1, 0]
x = np.random.uniform(0, 10, 50)
y = 2 * x + np.random.normal(0, 1, 50)
### ваш код здесь ###

ax.scatter(x, y, alpha=0.4, color="green", edgecolors='black', linewidth=0.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Scatter plot: y = 2x + шум', fontsize=12)
ax.grid(True, alpha=0.3)

# 4. Bar chart: точность моделей
ax = axes[1, 1]
models = ['A', 'B', 'C', 'D', 'E']
accuracy = [0.85, 0.92, 0.88, 0.95, 0.90]
### ваш код здесь ###

bars = ax.bar(models, accuracy, color="lightblue")
ax.set_xlabel('Модель')
ax.set_ylabel('Accuracy')
ax.set_title('Сравнение моделей', fontsize=12)
ax.grid(True, alpha=0.3)
ax.set_ylim(0.8, 1)
# Добавляем значения над столбцами
for bar, acc in zip(bars, accuracy):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.005,
            f'{acc:.2f}', ha='center')

plt.tight_layout()
plt.show()
```

```python
# Используйте датасет tips из seaborn:
# import seaborn as sns
# tips = sns.load_dataset('tips')

# Создайте 3 графика:
# 1. Box plot: total_bill по дням недели (day)
#    Разделите дополнительно по времени (time) с помощью hue
# 2. Violin plot: tip в зависимости от размера компании (size)
# 3. Pair plot: выберите 4 числовых признака, раскрасьте по полу (sex)

import seaborn as sns
import matplotlib.pyplot as plt

# Загружаем датасет
tips = sns.load_dataset('tips')

# Создаем фигуру для графиков
tips = sns.load_dataset('tips')
sns.set_style("whitegrid")

# 1. Box plot: total_bill по дням с разделением по времени
# data=tips, x='day', y='total_bill', hue='time'
plt.figure(figsize=(12, 10))
plt.subplot(2, 2, 1)
total_bill = ["Thurthday", "Friday", "Saturday", "Sunday"]
time = ["Dinner", "Lunch"]
sns.boxplot( data=tips, x='day', y='total_bill', hue='time')
plt.title('Box plot по категориям\n(сравнение распределений)')
plt.xlabel('День недели')
plt.ylabel('Total Bill')

# 2. Violin plot: tip по размеру компании
# x='size', y='tip'
plt.subplot(2, 2, 2)
sns.violinplot(data=tips, x='size', y='tip', color="steelblue", inner='box')
plt.title('Tip по размеру компании')
plt.xlabel('Размер компании')
plt.ylabel('Tip')

# 3. Pair plot: 4 числовых признака, раскраска по полу
# tips[['total_bill', 'tip', 'size', 'sex']]
# Создаем отдельную фигуру для pair plot
plt.figure(figsize=(10, 8))
sns.pairplot(
    tips[['total_bill', 'tip', 'size', 'sex']],
    hue='sex',
    palette='tab10',
    diag_kind='hist',  # столбики
    plot_kws={'alpha': 0.6}
)
plt.suptitle('Pair plot по полу', y=1.02)
plt.show()
```
